#**Exploring and Visualizing the Iris Dataset**

#**Objective**

####Understand the structure, trends, and distributions of the Iris dataset using basic data analysis and visualization.

##Step **1: Load Dataset**

```python
# pandas is used for data handling and analysis
import pandas as pd

# seaborn is used to load dataset and create plots
import seaborn as sns

# matplotlib is used to display graphs
import matplotlib.pyplot as plt

# Load Iris dataset from seaborn
# Dataset contains flower measurements and species
iris = sns.load_dataset("iris")

```

##**Step 2: Inspect Data**

```python
# Display total rows and columns
print("Shape of dataset:", iris.shape)

# Display column names
print("Column names:", iris.columns)

# Display first 5 records
print(iris.head())
```

##Step **3: Summary Statistics**

```python
# Display data types and missing values
iris.info()

# Display statistical summary of dataset
iris.describe()
```

##**Step 4: Visualizations**

###**Scatter Plot**

```python
# Scatter plot to visualize relationship between sepal length and width
# Different colors represent different species
sns.scatterplot(
    data=iris,
    x="sepal_length",
    y="sepal_width",
    hue="species"
)

# Show plot
plt.show()
```

###**Histograms**

```python
# Histogram to visualize distribution of numerical features
iris.hist(figsize=(10, 6))

# Show histograms
plt.show()
```

###**Box Plots**

```python
# Box plot to detect outliers
sns.boxplot(data=iris)

# Show box plot
plt.show()
```
